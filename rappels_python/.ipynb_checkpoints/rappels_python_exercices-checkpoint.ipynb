# Exercice 0.1

Écrire un code qui trouve les **n** nombres les plus grands dans une liste donnée.


Faire tourner ce code sur la liste [41,23,63,87,43,21] avec **n=2**. Est ce que ce code trouve bien 87 et 63 ?

Sans utiliser les fonction dans python on peut y arriver avec des boucles et des conditions:

```python
liste =  [41, 23, 63, 87, 43, 21]
n = 2

def maxListe(liste, nbMax):
    if nbMax > len(liste):
        print("Erreur paramètre")
    max = []
    i = 0
    liste.sort(reverse=True)
    for i in range(nbMax):
        max.append(liste[i])
    return max

print(maxListe(liste, n))
```

# Exercice 0.2

Écrire un script qui implemente un test permettant de savoir si **« z »**, un nombre entier, est divisible par 2, par 3 ou par 5.

Préciser aussi si **« z »** n’est divisible ni par 2, ni par 3, ni par 5.

Tester avec 24, 35, 49 et 60.


```python
def main(z):
    result = ""
    if z%2 == 0:
        result+="2 "
    if z%3 == 0:
        result+="3 "
    if z%5 == 0:
        result+="5 "
    if z%5 != 0 and z%3 != 0 and z%2 != 0:
        result+= "rien"
    return result

for z in [24, 35, 49, 60]:
    print(str(z) + " est divisible par " + main(z))
```

# Exercice 0.3
Écrire un code permettant de déterminer les racines d’un polynôme du second degré ax2+bx+c

On fournira les 3 nombres **« a »**, **« b »** , **« c »** et on calculera le discriminant **« delta »**. Selon la valeur négative, nulle ou positive de « delta », calculer et afficher les racines réelles du trinôme : « x12 » si racine double, « x1 » et « x2 » si racines distinctes. Attention au cas « a = 0 » ...

Essayer avec :
- (a=10, b=20, c=30)
- (a=4, b=12, c=9)
- (a=1, b=16, c=4)
- (a=0, b=1, c=−1)

```python
import math

def sqrt_polynomial(a, b, c):
    """
    Cette fonction calcule la racine carrée d'un polynôme de degré 2
    :param a: Coefficient de x^2
    :param b: Coefficient de x
    :param c: Terme constant
    :return: Racine carrée du polynôme
    """
    x1 = 0
    x2 = 0
    delta = b**2 - 4*a*c
    
    if delta > 0:
        x1 = (-b - np.sqrt(delta)) / 2*a
        x2 = (-b + np.sqrt(delta)) / 2*a
        return f"2 racines réelles : x1 = {x1} et x2 = {x2}"
    elif delta == 0:
        x1 = -b / 2*a
        return f"1 racines réelles : x = {x1}"
    else :
        return "pas de racines réelles (racines complexe)"

params = [10, 20, 30]
print(sqrt_polynomial(*params)) # equivalent à sqrt_polynomial(params[0], params[1], params[2])
```

Écrire **params** dans la parentèse permet d'appeller tous les éléments de l'objet params. Params prendra successivement les valeurs présente dans la liste **parameters**
Notez qu'il s'agits de tuple. En l'occurence, on aurait pu avoir des listes, cela ne change rien dans notre cas.

# Exercice 0.4


Écrire un script qui parcourt la liste 'objets_ss' et qui indique si l’objet est une planète tellurique (c’est-à-dire si l’objet fait partie de la liste Mercure, Venus, Terre, Mars), ou bien gazeuse (l’objet fait partie de la liste Jupiter, Saturne, Uranus, Neptune); sinon, indiquer que l’objet est une planète naine et donner leur nombre.


```python
def test_planete(planete):
    tellurique = ["Mercure", "Venus", "Terre", "Mars"]
    gazeuse = ["Jupiter", "Saturne", "Uranus", "Neptune"]
    if planete in tellurique:
        return "tellurique"
    elif planete in gazeuse:
        return "gazeuse"
    return "naine (parmis les 5 existantes)"

# Test        
objets_ss = np.array(['Neptune', 'Terre', 'Hauméa', 'Saturne', 'Cérès', 'Uranus',
       'Mercure', 'Venus', 'Éris', 'Mars', 'Jupiter', 'Makémaké',
       'Pluton'])

for planete in objets_ss:
    print(str(planete) + " est une planète : " + test_planete(planete))
```
